```
for t in df_1['Name of Table']:
  html = wp.page(t).html().encode("UTF-8")
  df1 = pd.read_html(html)[0] 
  
  lst.append(df1.head(5))
```

```
temp1=[]
```

```
from jinja2 import Template
```

```
for i in range(len(lst)):
  try:
    rank1=lst[i]['Rank']
    #rank=df1['Alexa top 50 global sites(As of April\xa016, 2020[3]']
    superlative=df_wiki['superlatives'][i]
    entity_class=df_wiki['Root'][i]
    #ec=''+ entity_class +''
    #entity=lst[i]['Sovereign state/dependency']
    entity=lst[i][df_wiki['Member'][i]]
    measure=df_wiki['Measure '][i]
    #df=df1.head(5)
    for index, row in lst[i].iterrows(): 
      rank=row['Rank']
      entity=row[df_wiki['Member'][i]]
      #entity=row['Sovereign state/dependency']
      tm=Template("{{entity}} is the {{rank}} {{superlative}} {{entity_class}}")
      #tm=Template("{{entity}} is the {{rank}} {{superlative}} {{entity_class}} by {{measure}}")
      msg = tm.render(entity=entity,rank=rank, superlative=superlative, entity_class=entity_class)
      temp1.append(msg)
      print(msg)

  except:
    print(i)
```

```
tempdf = pd.DataFrame(temp1,columns=['Template'])
```
